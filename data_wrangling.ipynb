Structuring

```python
#converting the 'date_added' to datetime
df['data_added']=pd.to_datetime(df['date_added'],format='mixed')
```

```python
# Separating 'duration' into numeric value and unit
df[['duration_value', 'duration_unit']] = df['duration'].str.extract(r'(\d+)\s*(\w+)')
```

```python
# Converting duration_value to numeric
df['duration_value'] = pd.to_numeric(df['duration_value'])
```

```python
# View Resulting columns
print(df[['duration_value', 'duration_unit']])
```

Cleaning

```python
#handling the duplicated by dropping them
df.drop_duplicates()
```

```python
# Create a new column to combine director and cast
df['dir_cast'] = df['director'].astype(str) + '---' + df['cast'].astype(str)

# Count the frequency of director-cast combinations
counts = df['dir_cast'].value_counts()

# Filter for combinations that appear 3 or more times
filtered_counts = counts[counts >= 3]
filtered_values = filtered_counts.index

# Create a mapping dictionary of director:cast
dict_direcast = {}
for item in filtered_values:
    try:
        director, cast = item.split('---')
        dict_direcast[director.strip()] = cast.strip()
    except:
        continue  # In case there's a malformed entry

# Fill missing director values based on matching cast
for director, cast in dict_direcast.items():
    df.loc[(df['director'].isna()) & (df['cast'] == cast), 'director'] = director
```

```python
df
```

```python
# Assign Not Given to all other director fields
df.loc[df['director'].isna(),'director'] = 'Nan'

#Use directors to fill missing countries
directors = df['director']
countries = df['country']
#pair each director with their country use zip() to get an iterator of tuples
pairs = zip(directors, countries)
# Convert the list of tuples into a dictionary
dir_cntry = dict(list(pairs))
```

```python
dir_cntry
```

```python
# Director matched to Country values used to fill in null country values
for i in range(len(dir_cntry)):
  df.loc[(df['country'].isna()) & (df['director'] == list(dir_cntry.items())[i][0]),'country'] = list(dir_cntry.items())[i][1]
```

```python
# Assign Not Given to all other country fields
df.loc[df['country'].isna(),'country'] = 'Nan'

# Assign Not Given to all other fields
df.loc[df['cast'].isna(),'cast'] = 'Nan'

# dropping other row records that are null
df.drop(df[df['date_added'].isna()].index,axis=0,inplace=True)
df.drop(df[df['rating'].isna()].index,axis=0,inplace=True)
df.drop(df[df['duration'].isna()].index,axis=0,inplace=True)
```

```python
#Removing any columns i  added during wrangling
df.drop(columns=['dir_cast'], inplace=True)
```

```python
# Save as CSV
df.to_csv('cleaned_netflix.csv', index=False)
```
